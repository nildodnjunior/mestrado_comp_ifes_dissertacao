```python
from pyalex import Works
from itertools import chain
```

```python
query = Works().search('("artificial intelligence" AND "active learning")')\
    .filter(publication_year=">2018", type='article', language='en')\
    .sample(10000)
```

```python
for record in chain(*query.paginate(per_page=100)):
    print(record['title'])
```
